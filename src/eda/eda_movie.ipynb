```python
actor_counts = (
    df.assign(actor=df["actors"].str.split('|'))
      .explode('actor')['actor']
      .value_counts()
      .head(20)      # chỉ lấy top 20 cho dễ nhìn
)

plt.figure(figsize=(12,6))
actor_counts.plot(kind='bar')
plt.title("Top Actors Appearing in Watched Movies")
plt.xlabel("Actor")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
df_actor = df.assign(actor=df['actors'].str.split('|')).explode('actor')

# total phim mỗi user
user_total = df.groupby('user_id')['movie_id'].nunique()

# đếm actor
user_actor_count = (
    df_actor.groupby(['user_id', 'actor'])['movie_id']
            .nunique()
            .reset_index(name='count')
)

# tính %
user_actor_count['pct'] = user_actor_count.apply(
    lambda row: row['count'] / user_total[row['user_id']],
    axis=1
)

# lấy diễn viên có % cao nhất mỗi user
user_actor_top = (
    user_actor_count.loc[user_actor_count.groupby('user_id')['pct'].idxmax()]
    [['user_id', 'actor', 'pct']]
    .rename(columns={'actor':'top_actor'})
)

user_actor_top.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(user_actor_top['pct'].values, marker='o', linestyle='-')
plt.title("Top Actor Similarity per User")
plt.xlabel("User Index")
plt.ylabel("Percentage")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

pct = user_actor_top['pct'].values
movie_count = user_total.loc[user_actor_top['user_id']].values

movie_count_norm = (movie_count - movie_count.min()) / (movie_count.max() - movie_count.min())

plt.figure(figsize=(14,6))

for i, y in enumerate(pct):
    plt.plot(i, y,
             marker='o',
             color='blue',
             alpha=movie_count_norm[i] * 0.8 + 0.2)   # đậm nhạt theo số phim

plt.title("Top Actor Similarity per User (Alpha = Movie Count)")
plt.xlabel("User Index")
plt.ylabel("Percentage")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ===== 1. Tổng số phim mỗi user =====
user_total = df.groupby('user_id')['movie_id'].nunique()

# ===== 2. Đếm số lần xem theo khung giờ cho mỗi user =====
user_hour_count = (
    df.groupby(['user_id', 'hour_watched'])['movie_id']
      .nunique()
      .reset_index(name='count')
)

# ===== 3. Tính % =====
user_hour_count['pct'] = user_hour_count.apply(
    lambda row: row['count'] / user_total[row['user_id']],
    axis=1
)

# ===== 4. Lấy khung giờ mà user xem nhiều nhất =====
user_hour_top = (
    user_hour_count.loc[user_hour_count.groupby('user_id')['pct'].idxmax()]
    [['user_id', 'hour_watched', 'pct']]
).reset_index(drop=True)

# ===== 5. Lấy số phim đã xem để dùng cho màu =====
movie_count = user_total.loc[user_hour_top['user_id']].values

# ===== 6. Plot =====
plt.figure(figsize=(14,6))

scatter = plt.scatter(
    np.arange(len(user_hour_top)),
    user_hour_top['pct'],
    c=movie_count,
    s=np.clip(movie_count, 10, 250),
    cmap='viridis',
    alpha=0.8
)

plt.colorbar(scatter, label="Movies Watched")
plt.title("User Time Watching Habit — % of Movies Watched in Their Most Frequent Hour")
plt.xlabel("User Index")
plt.ylabel("Percentage")
plt.grid(True)
plt.tight_layout()
plt.show()
```
